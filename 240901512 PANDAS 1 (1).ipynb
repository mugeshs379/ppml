```python
import pandas as pd
#Empty DataFrame
df=pd.DataFrame()
print(df)
```

```python
emp=pd.Series(['Parker','John','william'])
id=pd.Series([102,107,109,114])
frame={'Emp':emp,'ID':id}

result=pd.DataFrame(frame)
print("\nseries to Data frame\n")
print(result)
```

```python
print("nAdding new coloumn:\n")
result['Age']=pd.Series([35,24,40,38])
print(result)
    
```

```python
print("nDeleting one coloumn:\n")
del result['Age']
print(result)
```

```python
print("\nExtracting the thirds row:\n")
print(result.loc[2])
```

```python
d2=pd.DataFrame([['Dale',123],['Mark',143]],columns=['emp','Id'])
print("\nAdding new row values:\n",pd.concat([result,d2]))
```
